```python
dic={'Gaussian_NB':gnb,'Multinomial_NB':mnb,'Bernoulli_NB':bnb,"LogisticRegression":lr,"SGDClassifier":sgd,"DecisionTreeClassifier":dtree,"LGBMClassifier":lgbm}

acc=[]
pre_s=[]
recall_s=[]
def all_clf(clf,X_train,y_train,X_test,y_test):
    dic[clf]
    model=dic[clf].fit(X_train,y_train)
    y_pred=model.predict(X_test)
    
    acc.append(accuracy_score(y_test,y_pred))
    pre_s.append(precision_score(y_test,y_pred))
    recall_s.append(recall_score(y_test,y_pred))
    print(pre_s)
    

   
```
